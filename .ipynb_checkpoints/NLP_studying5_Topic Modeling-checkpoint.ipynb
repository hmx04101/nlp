<h2>토픽 모델링(Topic Modeling)</h2>
<b>문서 집합에서 주제를 찾아내기 위한 기술</b><br>
<b>특정 무특정 주제에 관한 문서에서는 특정 단어가 자주 등장할 것이라는 직관을 기반</b><br>
<b>주로 잠재 의미 분석과 잠재 디리클레 할당 기법이 있음</b>

<h2>잠재 의미 분석(Latent Semantic Analysis)</h2>
<b>잠재 의미 분석(LSA)는 문석 색인 의미 검색에 사용, 잠재 의미 인덱싱(Latent Semantic Indexing, LSI)로도 알려짐</b><br>
<b>문서, 단어의 기반이 되는 잠재적인 토픽을 발견하는 것</b><br>
<b>문서 모음에서 생성한 문서-단어 행렬(Document Term Matrix)에서 <br>1. 단어-토픽 행렬(Term-Topic Matrix)와<br>2. 토픽-중요도 행렬(Topic-Importance Matrix), 그리고 <br>3. 토픽-문서 행렬(Topic-Document Matrix)로 분해</b>

<h2>잠재 디리클레 할당(Latent Dirichlet Allocation)</h2>
<b>대표적인 토픽 모델링 알고리즘 중 하나</b>
<b>1. 사용자가 토픽 개수를 지정해 알고리즘에 전달</b><br>
<b>2. 모든 단어들을 토픽 중 하나에 할당 </b><br>
<b>3. 모든 문서의 모든 단어에 대해 단어 w가 가정에 의거, p(t|d), p(w|t)에 따라 토픽 재할당 (자신만 잘못된 토픽에 할당, 다른 모든 단어는 올바른 토픽에 할당되었다 가정)</b><br>
<b>p(t|d) - 문서 d의 단어 중 토픽 t에 해당되는 비율, p(w|t) - 단어 w를 가지고 있는 모든 문서들 중 토픽 t가 할당된 비율 </b>

```python
from sklearn.datasets import fetch_20newsgroups
```

```python
dataset = fetch_20newsgroups(shuffle = True, random_state = 1,
                            remove = ('headers', 'footers', 'quotes'))
documents = dataset.data
print(len(documents))
documents[0]
```

```python
documents[3]
```

```python
import re
import nltk
from nltk.corpus import stopwords
from gensim.parsing.preprocessing import preprocess_string
```

```python
def clean_text(d):
    pattern = r'[^a-zA-Z\s]'
    text = re.sub(pattern, '', d)
    return d

def clean_stopword(d):
    stop_words = stopwords.words('english')
    return ' '.join([w.lower() for w in d.split() if w not in stop_words and len(w) > 3])

def preprocessing(d):
    return preprocess_string(d)
```

```python
import pandas as pd

news_df = pd.DataFrame({'article':documents})
print(len(news_df))
```

```python
news_df.replace("", float("NaN"), inplace = True)
news_df.dropna(inplace = True)
print(len(news_df))
```

```python
news_df['article'] = news_df['article'].apply(clean_text)
news_df['article']
```

```python
news_df['article'] = news_df['article'].apply(clean_stopword)
news_df['article']
```

```python
tokenized_news = news_df['article'].apply(preprocessing)
tokenized_news = tokenized_news.to_list()
tokenized_news[0]
```

```python
import numpy as np

print(len(news_df))
drop_news = [index for index, sentence in enumerate(tokenized_news) if len(sentence) <= 1]
news_texts = np.delete(tokenized_news, drop_news, axis = 0)
print(len(news_texts))
```

<b>Gensim 이용한 토픽 모델링</b>

```python
from gensim import corpora
```

```python
dictionary = corpora.Dictionary(news_texts)
corpus = [dictionary.doc2bow(text) for text in news_texts]

print(corpus[1])
```

<b>잠재 의미 분석 LsiModel</b>

```python
from gensim.models import LsiModel
```

```python
lsi_model = LsiModel(corpus, num_topics = 20, id2word = dictionary)
topics = lsi_model.print_topics()
topics
```

```python
from gensim.models.coherencemodel import CoherenceModel
```

```python
min_topics, max_topics = 20, 25
coherence_score = []

for num_topics in range(min_topics, max_topics):
    model = LsiModel(corpus, num_topics = num_topics, id2word = dictionary)
    coherence = CoherenceModel(model = model,
                              texts = news_texts,
                              dictionary = dictionary)
    coherence_score.append(coherence.get_coherence())
    
print(coherence_score)
```

<b>잠재 리디클레 할당을 위한 LdaModel</b>

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

```python
x = [int(i) for i in range(min_topics, max_topics)]

plt.figure(figsize = (10, 6))
plt.plot(x, coherence_score)
plt.xlabel('Number of Topics')
plt.ylabel('Coherence')
```

```python
lsi_model = LsiModel(corpus, num_topics = 24, id2word = dictionary)
topics = lsi_model.print_topics(num_topics = 24)
topics
```

<b>잠재 디리클레 할당을 위한 LdaModel</b>

```python
from gensim.models import LdaModel
```

```python
lda_model = LdaModel(corpus, num_topics = 20, id2word = dictionary)
topics = lda_model.print_topics()
topics
```

```python
from gensim.models.coherencemodel import CoherenceModel
```

```python
min_topics, max_topics = 20, 25
coherence_score = []

for num_topics in range(min_topics, max_topics):
    model = LdaModel(corpus, num_topics = num_topics, id2word = dictionary)
    coherence = CoherenceModel(model = model,
                              texts = news_texts,
                              dictionary = dictionary)
    coherence_score.append(coherence.get_coherence())
    
print(coherence_score)
```

```python
x = [int(i) for i in range(min_topics, max_topics)]

plt.figure(figsize = (10, 6))
plt.plot(x, coherence_score)
plt.xlabel('Number of Topics')
plt.ylabel('Coherence')
```

```python
lda_model = LdaModel(corpus, num_topics = 24, id2word = dictionary)
topics = lda_model.print_topics()
topics
```
